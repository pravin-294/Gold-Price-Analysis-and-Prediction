```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 

df = pd.read_csv("../data/gold_price.csv")

df.head(5)

df.shape
df.columns
df.info() 
```

```python
# Create a new 'Date' column by combining Year and Month columns

# Convert Year column to string
# Convert Month column to string
# Combine them in 'YYYY-MM-01' format
# '01' is used as a placeholder day since we only have monthly data

df['Date'] = pd.to_datetime(
    df['Year'].astype(str) + '-' +
    df['Month'].astype(str) + '-01'
)
df.head()

```

```python
# Sort the data by Date to ensure correct chronological order
df.sort_values("Date")
df[['Year', 'Month', 'Date']].head()
```

```python
df['Date'].min()
df['Date'].max()
```
